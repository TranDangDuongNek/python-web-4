```python
class Student:
    def __init__(self, name):
        self.name = name
        self.grades = []

    def add_grade(self, grade):
        self.grades.append(grade)

    def calculate_average(self):
        if not self.grades:
            return 0
        return sum(self.grades) / len(self.grades)


class GradeManager:
    def __init__(self):
        self.students = []

    def add_student(self, name):
        if self.find_student(name):
            print(f"HS: '{name}' SPAM!")
        else:
            self.students.append(Student(name))
            print(f"+1 HA '{name}' JOIN SEVER.")

    def find_student(self, name):
        for student in self.students:
            if student.name == name:
                return student
        return None
# A.I =)) 
    def record_grade(self, name, grade):
        student = self.find_student(name)
        if student:
            student.add_grade(grade)
            print(f"Đã ghi điểm {grade} cho học viên '{name}'.")
        else:
            print(f"HS: '{name}' KO CÓ ")

    def calculate_average_all(self):
        if not self.students:
            print("Chưa có học viên nào.")
            return
        print("ĐIẾM TRUNG BÌNH HS")
        for student in self.students:
            avg = student.calculate_average()
            print(f"{student.name}: {avg:.2f} điểm")
# 

def main():
    manager = GradeManager()

    while True:
        print("\n----- CHƯƠNG TRÌNH QUẢN LÝ ĐIỂM -----")
        print("1. Thêm học viên mới")
        print("2. Ghi điểm cho học viên")
        print("3. Tính điểm trung bình của tất cả học viên")
        lua_chon = input("Nhập lựa chọn (1-5): ")

        if lua_chon == "1":
            name = input("Nhập tên học viên: ")
            manager.add_student(name)

        elif lua_chon == "2":
            name = input("Nhập tên học viên: ")
            try:
                grade = float(input("Nhập điểm: "))
                manager.record_grade(name, grade)
            except ValueError:
                print("Vui lòng nhập điểm hợp lệ (số).")

        elif lua_chon == "3":
            manager.calculate_average_all()

        else:
            print("HMM LỰA CHỌN NÀY ĐỆ HẠ KO ĐOÁN ĐC")

if __name__ == "__main__":
    main()
```
